```python
#Import necessary packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Number of Nuclei of Element A 
Na0=200.0
#Number of Nuclei of Element B
Nb0=5.0
#Time Constant of Element A
tauA=1.0
tauB=2.0
tauB1=(tauA/tauB)
#t0=1.0
```

```python
#Creates empty array of values for elements A and B
Na_t=[]
Nb_t=[]
#Adds the N0 values to the beginning of each array
Na_t.append(Na0)
Nb_t.append(Nb0)

#Exact analytical solutions
Na_exact = []
Na_exact.append(Na0)
Nb_exact = []
Nb_exact.append(Nb0)
```

```python
#Array containing the physical time
t0=0.0
tA=[]
tA.append(t0)
tB=[]
tB.append(t0)
```

```python
#Timestep interval
dt=0.01

#Maximum number of intervals
max_interval=1000
```

```python
#Here we implement the Euler Method to produce results showing the relationship between the number of nuclei 
#at a given time.

for i in range(1,max_interval):
    tmpA=Na_t[i-1]-Na_t[i-1]*dt
    tmpB=Nb_t[i-1]+((Na_t[i-1]-((Nb_t[i-1]*tauB1)))*dt)
    Na_t.append(tmpA)
    Nb_t.append(tmpB)
    
    #Now we convert our results from the previous cell back into a physical time rather than a dimensionless time.
    tmp = float(i)*dt*tauA
    tA.append(tmp)
    
    tmpB = float(i)*dt*tauB
    tB.append(tmpB)
    
    #Here we create and fill in the arrays containing the values computed by the analytical solutions. 
    tmp2 = Na0*np.exp(-tmp/tauA)
    Na_exact.append(tmp2)

    #tmp3=-(1/(-1+tauB1))*(Nb0*np.exp(-tmpB-(tmpB/tauB))*(np.exp(tmpB)-41*np.exp(tmpB)*tauB+40*np.exp(tmpB/tauB)*tauB))/(-1+tauB)
    tmp3=((tauB*Na0)/(tauA-tauB))*np.exp(-tmp/tauA)+(Nb0-((tauB*Na0)/(tauA-tauB)))*np.exp(-tmp/tauB)
    Nb_exact.append(tmp3)
```

```python
#Plot Figures
plt.figure(figsize=(5,5))
plt.plot(tA,Na_t,'y.',markersize=10.00, label='Numerical Solution')
plt.plot(tA,Na_exact,linestyle='-',color='k', label='Exact Solution')
plt.title("Decay of Element A Nuclei Over Time")
plt.legend()
plt.xlabel("Time")
plt.ylabel("N[A]")
```
